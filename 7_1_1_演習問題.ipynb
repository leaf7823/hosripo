```
# 7-1-1．演習問題

result = []

for i in range(1,101):
    if  i % 15 == 0:
        result.append(str('FizzBuzz!!'))
    elif i % 3 == 0:
        result.append('Fizz')
    elif i % 5 == 0:
        result.append('Buzz')
    else:
        result.append(i)  
result
```
