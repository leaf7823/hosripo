```
# スライスで文字列を抽出
#  flower
#  012345
#   f l o w e r
#  0 1 2 3 4 5 6
# FLOWER という文字列から LOW という部分文字列を取得する場合、開始インデックスに 
# 1 、終了インデックスに 4 を指定します。これは次のようにインデックスの位置を少しず
# らして考えてもらうと分かりやすいかと思います。

a = 'flower'
print(a[1:4])


s = 'abcde'

print(s[2:4])
# bc

print(s[:3])
# abc

print(s[1:])
# bcde
```

```
class TestClass:
    def __init__(self):
        self.val0 = 1
    
    def sum(self,val1,val2):
        sum = self.val0 + val1 + val2
        print(sum)
        
test_instance = TestClass()
test_instance.sum(4,5)
```

```
class TestClass1:
    def __init__(self):
        self.val0 = 1
    def sum1(self,val1,val2):
        sum = val1 + val2
        print(sum)
        
class TestClass2(TestClass1):
    def __init__(self):
        self.val3 = 5
    def sum2(self,val4,val5):
        sum = self.val3 + val4 + val5
        print(sum)
        
test_instance = TestClass2()
test_instance.sum1(1,2)
test_instance.sum2(1,2)
```

```
# fizzbuzz

for num in range(1, 10):
    if num % 3 == 0 and num % 5 == 0:   #　またはif num % 15 == 0:
       print("3と5の両方の倍数")
    elif num % 3 == 0:
       print(str(num) + " " + "3の倍数")
    elif num % 5 == 0:
       print(str(num) + " " + "5の倍数")
    else:
      print(num)
```
